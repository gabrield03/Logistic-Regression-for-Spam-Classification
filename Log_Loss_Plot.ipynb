```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/gabrield03/Logistic-Regression-for-Spam-Classification/main/Log_Loss_Iterations.csv'

loss = pd.read_csv(url)
loss
```

```python
plt.figure(figsize=(10, 6))

sns.set_palette('Set3')
sns.set_style('whitegrid')

sns.lineplot(
    data=loss,
    x='Iteration',
    y='Log Loss Value',
    linewidth=2.5,
)

plt.title('Log Loss by Iteration')
```
