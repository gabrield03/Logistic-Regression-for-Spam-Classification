```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
email_train_df = pd.read_csv('Data/train-1.csv')
email_train_df.shape
```

```python
email_test_df = pd.read_csv('Data/test-1.csv')
email_test_df.shape
```

```python
combined_df = pd.concat([email_train_df, email_test_df])
combined_df
```

```python
''' Select columns'''
# Assign the independent variable
feature_df = combined_df.iloc[:, :-1]
X = np.asarray(feature_df)
print(X)

y = np.asarray(combined_df['label'])
print(y)
```

```python
''' 
Divide the data into training and test data
if we didn't have the test-1.csv data, we would split the current dataset into train/test

from sklearn.model_select import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
'''

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

X_train.shape
y_train.shape

X_test.shape
y_test.shape
```

```python
''' Use the SVM model '''
from sklearn import svm

classifier = svm.SVC(kernel='linear', gamma='auto', C=2)

# fit the data
classifier.fit(X_train, y_train)

# Output is 1D
y_predict = classifier.predict(X_test)
```

```python
''' Evaluate the results '''
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```

```python
''' Comparison with Logistic Regression '''
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1e5)
logreg.fit(X_train, y_train)

y_predict = logreg.predict(X_test)
```

```python
''' Evaluate the results '''
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```
